```python
titles = {lang:[x[2] for x in urns[lang]] for lang in urns}
```

```python
years = {lang:[x[3] for x in urns[lang]] for lang in urns}
```

```python
from collections import Counter
```

```python
tot = dict()
for lang in urns:
    tot[lang] = Counter()
    for t in titles[lang]:
        tot[lang].update(t.split())
```

```python
tst = Counter()
tst.update(years['nob'])
tst.most_common(20)
```

```python
anno = dict()
for lang in urns:
    anno[lang] = Counter()
    anno[lang].update(years[lang])
```
